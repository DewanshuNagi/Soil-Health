```python
import pandas as pd

# Identify numeric columns
numeric_cols = df.select_dtypes(include='number').columns

# Function to remove outliers using IQR
def remove_outliers_iqr(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Apply the function
df_no_outliers = remove_outliers_iqr(df, numeric_cols)

# Save the cleaned data
df_no_outliers.to_csv("soil_health_no_outliers.csv", index=False)

print("Outliers removed. New dataset saved as: soil_health_no_outliers.csv")
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
# Ensure output folder exists
output_folder = "eda_outputs"
os.makedirs(output_folder, exist_ok=True)
```

```python
# 1. Summary statistics for numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
summary_stats = df[numeric_cols].describe()
summary_stats.to_csv(os.path.join(output_folder, "summary_statistics.csv"))
summary_stats
```

```python
# 2. Correlation heatmap
plt.figure(figsize=(14, 10))
corr = df[numeric_cols].corr()
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt='.2f')
plt.title("Correlation Heatmap of Soil Nutrients")
plt.tight_layout()
plt.show()
```

```python
# 3. Average deficiency by state
deficient_cols = [col for col in numeric_cols if 'low' in col or 'deficient' in col]
avg_deficiency_by_state = df.groupby("state")[deficient_cols].mean()
avg_deficiency_by_state.to_csv(os.path.join(output_folder, "avg_deficiency_by_state.csv"))
avg_deficiency_by_state
```

```python
# 4. Top 5 nitrogen-deficient states
top_nitrogen_states = avg_deficiency_by_state['nitrogen - low'].sort_values(ascending=False).head(5)

# Save to CSV
top_nitrogen_states.to_csv(os.path.join(output_folder, "top_nitrogen_deficient_states.csv"))

# Convert to DataFrame for display
top_nitrogen_states_df = top_nitrogen_states.reset_index()
top_nitrogen_states_df.columns = ['State', 'Nitrogen - Low']

# Display the DataFrame
print(top_nitrogen_states_df)
```

```python
# 5. Boxplot of nitrogen-low in top 5 states
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[df['state'].isin(top_nitrogen_states.index)], x='state', y='nitrogen - low')
plt.title("Nitrogen Deficiency (Low) in Top 5 Affected States")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6. Heatmap of nutrient deficiency by state
plt.figure(figsize=(16, 10))
sns.heatmap(avg_deficiency_by_state, cmap='Reds', linewidths=0.5)
plt.title("Average Nutrient Deficiency by State")
plt.tight_layout()
plt.show()
```

```python

# 1. Group by year for selected nutrients
yearly_avg = df.groupby("year")[[
    "nitrogen - low", "phosphorous - low", "potassium - low",
    "zn - deficient", "boron - deficient"
]].mean().reset_index()

# 2. Line plot for each nutrient
plt.figure(figsize=(12, 6))
for col in yearly_avg.columns[1:]:
    plt.plot(yearly_avg["year"], yearly_avg[col], marker='o', label=col)

plt.title("Year-by-Year Nutrient Deficiency Comparison")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "yearly_nutrient_trends.png"))
plt.close()

# 3. Heatmap of nutrient deficiency by year
plt.figure(figsize=(10, 6))
sns.heatmap(yearly_avg.set_index("year"), annot=True, cmap="YlOrRd", fmt=".1f", linewidths=0.5)
plt.title("Heatmap of Nutrient Deficiency by Year")
plt.tight_layout()
plt.show()
```
